# K Means - Generated data

Isotropic dataset module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html) \
Train-test splits in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.train_test_split.html) \
k-Means module in sklearn [here](https://scikit-learn.org/stable/modules/generated/sklearn.cluster.KMeans.html)

```
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
```

```
X, _ = make_blobs(n_samples = 500, n_features = 2, random_state = 170)

# Create train (80%) and test (20%) sets and print all shapes
X_train, X_test = train_test_split(X, test_size= 0.2, random_state = 170)
print(X_train.shape, X_test.shape)

_, ax = plt.subplots(figsize=(5,4))
ax.scatter( X_train[:,0], X_train[:,1], edgecolors='k', c = 'b')
ax.scatter(X_test[:,0], X_test[:,1], edgecolors = 'k', c = 'r')
```

```
kmeans = KMeans(n_clusters=3)
# Train Kmeans
kmeans.fit(X_train) #(n_samples, n_features)
```

```
# Plot centroids
centroids = kmeans.cluster_centers_
print(centroids)
```

```
# Create predictions
predictions = kmeans.predict(X_test)
print(predictions.shape)
```

```
_, ax = plt.subplots (figsize = (5,4))
colors=['blue','cyan','magenta']

# plot data and found centroids
ax.scatter(X_test[:,0], X_test[:,1], c = predictions, edgecolors = 'k', cmap = 'Paired')
ax.scatter(centroids[:,0], centroids[:,1], c = colors, marker = 'o', s = 150)
```
